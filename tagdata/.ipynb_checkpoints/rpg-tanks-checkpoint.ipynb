```python
#tank_level_data = [df.columns.values.tolist()] + df.values.tolist()
tank_level_data = df.values.tolist()
# for data in tank_level_data:
time = []
WE1FT_L = []
ILWT1_L = []
TAL_1076_DO = []
SJ_2744_DO = []

for data in tank_level_data:
    time.append(data[0])
    WE1FT_L.append(data[1])
    ILWT1_L.append(data[2])
    TAL_1076_DO.append(data[3])
    SJ_2744_DO.append(data[4])
# time

def checkError():
    for j in range(0,len(time)-1):
        a = WE1FT_L[j+1] - WE1FT_L[j]
        b = ILWT1_L[j+1] - ILWT1_L[j]
        if (TAL_1076_DO[j]==0 and SJ_2744_DO[j]==0 and (a!=0 or b!=0)):
            print("Abnormal Transfer taking Place at time: "+time[j])
        
checkError()    
```
